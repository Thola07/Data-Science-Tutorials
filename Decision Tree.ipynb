```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import pylab as pl
%matplotlib inline
from sklearn.tree import DecisionTreeClassifier
from sklearn import preprocessing
```

```python
df = pd.read_csv("~\Desktop\iris.csv")
```

```python
X = df[["sepal_length","sepal_width","petal_length","petal_width"]].values
X[0:5]
```

```python
y = df[["species"]]
y[0:5]
```

```python
from sklearn.model_selection import train_test_split
X_trainset, X_testset, y_trainset,  y_testset = train_test_split(X, y, test_size = 0.2 , random_state = 0)
print("Train set:", X_trainset.shape, y_trainset.shape)
print("Test set:", X_testset.shape, y_testset.shape)
```

```python
irisTree = DecisionTreeClassifier(criterion = "entropy", max_depth = 2)
irisTree
```

```python
irisTree.fit(X_trainset,y_trainset)
```

```python
predTree = irisTree.predict(X_testset)
```

```python
print(predTree [0:5])
print(y_testset [0:5])
```

```python
from sklearn import metrics
import matplotlib.pyplot as plt
print("DecisionTrees's Accuracy: ", metrics.accuracy_score(y_testset, predTree))
```
